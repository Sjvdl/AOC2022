```python
import pandas as pd

data = pd.read_csv("day8.txt", header=None)
data = data[0].str.split("", expand=True)

dt = data.drop(data.columns[0], axis=1)
dt2 = dt.drop(data.columns[100], axis=1)
```

```python
with open("day8.txt") as f:
    text = f.read().splitlines()
    
revlines = []
for line in text:
    linelen = len(line)
    revline = line[linelen::-1]
    revlines.append(revline)
    
hor_mask = []
for line in text:
    hor_line_mask = []
    highest = -1
    for char in line:
        height = int(char)
        if height > highest:
            visible = True
            highest = height
        else:
            visible = False
        hor_line_mask.append(visible)
    hor_mask.append(hor_line_mask)
    
vert_mask = []
for nr in range(0, len(text[0])):
    vert_line_mask = []
    highest = -1
    for line in text:
        height = int(line[nr])
        if height > highest:
            visible = True
            highest = height
        else:
            visible = False
        vert_line_mask.append(visible)
    vert_mask.append(vert_line_mask)
    
rev_hor_mask = []
for line in revlines:
    rev_hor_line_mask = []
    highest = -1
    for char in line:
        height = int(char)
        if height > highest:
            visible = True
            highest = height
        else:
            visible = False
        rev_hor_line_mask.append(visible)
    rev_hor_mask.append(rev_hor_line_mask)
    
rev_vert_mask = []
for nr in range(0, len(revlines[0])):
    rev_vert_line_mask = []
    highest = -1
    revrevlines = revlines.copy()
    revrevlines.reverse()
    for line in revrevlines:
        height = int(line[nr])
        if height > highest:
            visible = True
            highest = height
        else:
            visible = False
        rev_vert_line_mask.append(visible)
    rev_vert_line_mask.reverse()
    rev_vert_mask.append(rev_vert_line_mask)
```

```python
trans_hor_mask = list(map(list, zip(*hor_mask)))
trans_rev_hor_mask = list(map(list, zip(*rev_hor_mask)))
```

```python
dtb = dt2.copy()
display(dtb)

nr = 0
for s in dtb:
    m1 = pd.Series(trans_hor_mask[nr])
    m2 = pd.Series(vert_mask[nr])
    dtb.loc[m1 | m2, s] = True
    #dtb[s].mask(m1 | m2, True, inplace=True)
    nr += 1

nr2 = 0
for s in dtb:
    m3 = pd.Series(trans_rev_hor_mask[nr2])
    m4 = pd.Series(rev_vert_mask[nr2])
    col = 100 -s
    dtb.loc[m3 | m4, col] = True
    #dtb[col].mask(m3 | m4, True, inplace=True)
    nr2 += 1
    
visible_trees = 0

tfdtb = dtb.replace(r"^[0-9]$", value=False, regex=True)
display(tfdtb)

true_count = tfdtb.sum()
visible_trees = true_count.sum()
```
